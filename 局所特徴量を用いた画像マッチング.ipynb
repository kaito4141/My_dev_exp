```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import ImageGrid
%matplotlib inline
import math
img1 = cv2.imread('/content/drive/MyDrive/pro/img/test_img0101.jpg')
img2 = cv2.imread('/content/drive/MyDrive/pro/img/test_img0102.jpg')
img_gray1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img_gray2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
orb1 = cv2.ORB_create()
ak1=cv2.AKAZE_create()
kp1, des1 = orb1.detectAndCompute(img_gray1,None)
kp01, des01 = ak1.detectAndCompute(img_gray1,None)
ak2=cv2.AKAZE_create()
orb2 = cv2.ORB_create()
kp2, des2 = orb2.detectAndCompute(img_gray2,None)
kp02, des02 = ak2.detectAndCompute(img_gray2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,True)
matches1 = bf.match(des1, des2)
matches2=bf.match(des01,des02)
h1,w1=img_gray1.shape
f1_img = np.zeros((h1,w1), np.uint8)
```

```python
f1_img = cv2.drawMatches(img1, kp1,
img2, kp2, matches1,None,0)
f2_img=cv2.drawMatches(img1, kp01,
img2, kp02, matches2,None,0)
```

```python
plt.imshow(cv2.cvtColor(f1_img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(f2_img, cv2.COLOR_BGR2RGB))
```
